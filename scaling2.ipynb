### C. _Normalization_

- aka Minimum Maximum Scaler

- $\displaystyle x_\textrm{normalized} = \frac {x - x_\textrm{min}} {x_\textrm{max} - x_\textrm{min}}$

- Normalization result range is between  __0__ - __1__ 

- Inverse Norm $\displaystyle x = x_\textrm{normalized} \cdot ({x_\textrm{max} - x_\textrm{min}}) + x_\textrm{min}$

````python
df = pd.DataFrame({
    'TB': [1.72, 1.81, 1.93, 1.67, 1.85, 1.66, 1.59, 1.76, 1.88, 1.78],
    'BB': [76, 65, 87, 55, 60, 78, 98, 77, 74, 64]
})
df
````

````python
df['TB norm'] = (df['TB'] - df['TB'].min()) / (df['TB'].max() - df['TB'].min())
df['BB norm'] = (df['BB'] - df['BB'].min()) / (df['BB'].max() - df['BB'].min())
df
````

<hr>

### D. _Normalization on Sklearn_


#### D1. _```minmax_scale()```_

``` from sklearn.preprocessing import minmax_scale```

````python
from sklearn.preprocessing import minmax_scale
````

````python
df['TB norm 2'] = minmax_scale(df['TB'])
df['BB norm 2'] = minmax_scale(df['BB'])
df
````

#### D2. _```MinMaxScaler()```_
```from sklearn.preprocessing import MinMaxScaler```

````python
from sklearn.preprocessing import MinMaxScaler
````

````python
scaler = MinMaxScaler()
scaler.fit(df[['TB', 'BB']])

N3 = scaler.transform(df[['TB', 'BB']])

df['TB norm 3'] = N3[:,0]
df['BB norm 3'] = N3[:,1]
df
````

````python
in3 = scaler.inverse_transform(df[['TB norm 3', 'BB norm 3']])
df['TB inv'] = in3[:,0]
df['BB inv'] = in3[:,1]
df
````

````python
### apakah normalisasi merubah sebaran data?
````

````python
plt.subplot(121)
plt.plot(df['TB'], df['BB'], 'ro')
plt.subplot(122)
plt.plot(df['TB norm'], df['BB norm'], 'go')
plt.show()
````

### CASE

````python
a = pd.DataFrame({
    'mesin': [1000, 2000, 3000, 4000, 5000],
    'harga': [10, 25, 35, 55, 80]
})
````

````python
scaler = MinMaxScaler()
scaler.fit_transform(a[['mesin']])
````

````python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model2 = LinearRegression()

model1.fit(a[['mesin']], a['harga'])
model2.fit(scaler.fit_transform(a[['mesin']]), a['harga'])
````

````python
a['harga pred 1'] = model1.predict(a[['mesin']])
a['harga pred 2'] = model2.predict(scaler.fit_transform(a[['mesin']]))
a
````
